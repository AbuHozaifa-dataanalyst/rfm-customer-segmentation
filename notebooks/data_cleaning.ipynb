# Retail Sales Data Cleaning for RFM Analysis

## Load Data



```python
import pandas as pd

df = pd.read_csv("../data/raw/online_retail_II.csv", encoding='latin1')

df.to_csv("../data/raw/online_retail_II.csv.gz", index=False, compression='gzip')
df.head(5)
```

## Check Missing Values

```python
df.isnull().sum()
```

## Remove Missing Customer IDs

```python
df = df.dropna(subset=['Customer ID'])
```

## Remove Duplicates

```python
df = df.drop_duplicates()
```

## Remove Negative Quantity

```python
df = df[df['Quantity'] > 0]
```

## Remove Invalid Prices

```python
df = df[df['Price'] > 0]
```

## Fix Date Foremat

```python
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
```

## Create Total Amount

```python
df['TotalAmount'] = df['Quantity'] * df['Price']
```

## Rename Columns

```python
df = df.rename(columns={
    'Customer ID': 'customer_id',
    'Invoice': 'invoice_no',
    'InvoiceDate': 'invoice_date',
    'Quantity': 'quantity',
    'Price': 'unit_price'
})
```

## Final Data Check

```python
df.info()
df.describe()
```

## Save Clean Data

```python
df.to_csv('../data/processed/clean_retail_data.csv.gz', index=False, compression='gzip')
```
